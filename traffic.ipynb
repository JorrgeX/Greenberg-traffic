```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
# Define functions
def f(p):
    return -p * np.log(p)

def df(p):
    return -np.log(p) - 1
```

```python
fig = plt.figure()

s = (f(0.1) - f(0.9)) / (0.1 - 0.9)

t = np.array([0, 1])
plt.plot([0, s], t, 'k--', linewidth=2)

for i in range(1, 41):
    # left characteristic
    if -i / 5 + df(0.1) > s:
        tintersect = -i / 5 / (s - df(0.1))
        plt.plot([-i / 5, -i / 5 + df(0.1) * tintersect], [0, tintersect], 'k')
    else:            
        plt.plot([-i / 5, -i / 5 + df(0.1)], t, 'k')

    # right characteristic
    if i / 10 + df(0.9) < s:
        tintersect = i / 10 / (s - df(0.9))
        plt.plot([i / 10, i / 10 + df(0.9) * tintersect], [0, tintersect], 'k')
    else:
        plt.plot([i / 10, i / 10 + df(0.9)], t, 'k')

plt.xlabel('x')
plt.ylabel('t')
plt.xlim([-1, 1])
plt.ylim([0, 1])
plt.show()
```

```python
xs = fsolve(lambda x: 1 + (5 * np.pi / 9 + np.arctan(x)) * 2 * x, 10)
Tb = (5 * np.pi / 9 + np.arctan(xs)) * (1 + xs**2)

p0 = lambda x: 0.5 + 0.9 / np.pi * np.arctan(x)
plt.figure()
for x0 in np.linspace(-10, 10, 200):
    slope = df(p0(x0))
    plt.plot([x0, x0 + slope * 2], [0, 2], 'k')

plt.plot([-1, 1], [Tb, Tb], 'k--', linewidth=2)
plt.xlim([-1, 1])
plt.show()

s = (f(0.05) - f(0.95)) / (0.05 - 0.95)
print("s =", s)
print("Tb =", Tb)
```
